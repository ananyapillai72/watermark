```python
import cv2
import numpy as np
import requests
from PIL import Image

# Load the image and logo
image = Image.open(requests.get('https://images.unsplash.com/photo-1597848212624-a19eb35e2651?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxleHBsb3JlLWZlZWR8MXx8fGVufDB8fHx8&w=1000&q=80', stream=True).raw)
logo = Image.open(requests.get('https://cdn.dribbble.com/users/2926335/screenshots/6767047/logo.png', stream=True).raw)

# Resize the image and logo
img_logo = image.resize((300, 500))
img_text = image.resize((300, 500))
logo = logo.resize((100, 100))

# Convert the images to numpy arrays
img_logo = np.array(img_logo.convert('RGB'))
h_img, w_img, _ = img_logo.shape
logo = np.array(logo.convert('RGB'))
h_logo, w_logo, _ = logo.shape

# Calculate the coordinates of the top-left corner of the logo
top_y = 0
left_x = 0
bottom_y = h_logo
right_x = w_logo

# Get the ROI of the image
roi = img_logo[top_y: bottom_y, left_x: right_x]

# Blend the logo with the ROI of the image
result = cv2.addWeighted(logo, 0.5, roi, 0.7, 0)

# Place the blended image back into the original image
img_logo[top_y: bottom_y, left_x: right_x] = result

# Convert the numpy array back into an image
img = Image.fromarray(img_logo, 'RGB')

# Display the image
img.show()
```
